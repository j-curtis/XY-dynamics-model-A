```python
import numpy as np
from matplotlib import pyplot as plt 
```

```python
thetas = np.load("thetas.npy")
```

```python
ntimes = thetas.shape[0]
L = int(np.sqrt(thetas.shape[1]))
```

```python
thetas.shape
```

```python
thetas = thetas.reshape(ntimes,L,L)
```

```python
thetas.shape
```

```python
plt.imshow(thetas[500,:,:])
plt.show()
plt.imshow(thetas[1000,:,:])
plt.show()
plt.imshow(thetas[1500,:,:])
plt.show()
```

```python
cx = np.exp(1.j*(thetas[1000:,:,0]-np.outer(thetas[1000:,0,0],np.ones(L))))
```

```python
plt.plot(np.abs(np.mean(cx,axis=0))**2)
plt.show()
```
